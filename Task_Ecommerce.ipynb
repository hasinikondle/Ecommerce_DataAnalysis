# Analyzing Ecommerce Sales Drop


**Company scenario(Flipcart/amazon)**:


we are seeing a drop in revenue find the reason and suggest actions


```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# load the data
customers = pd.read_csv("customers.csv")
products = pd.read_csv("products.csv")
orders = pd.read_csv("orders.csv")
```

```python
df = pd.merge(customers,orders,on="customer_id",how="inner")
df = pd.merge(df,products,on="product_id",how="inner")
df
```

**EDA**

```python
df.info()
```

```python
# converting object date coulumns into datetime datatype format
df["signup_date"] = pd.to_datetime(df["signup_date"])
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df.info()
```

```python
# create revenue column
# formula: revenue = price * quantity
df["revenue"] = df["price"]*df["quantity"]
df.head()
```

```python
# revenue trend by order_date
revenue = df.groupby("order_date")["revenue"].sum() #get revenue according to order_date
revenue = pd.DataFrame(revenue).reset_index() #convert series to dataframe
plt.plot(revenue["order_date"],revenue["revenue"],marker="o",mec="orange",mfc="orange") #plot graph to see revenue by order_date
plt.xlabel("Order date")
plt.ylabel("Revenue")
plt.title("Revenue by order date")
```

**The revenue is very less in the last two months(jun,jul)**

the drop is started from mid of may

```python
# revenue trend by city
revenue = df.groupby("city")["revenue"].sum() #get revenue according to order_date
revenue = pd.DataFrame(revenue).reset_index() #convert series to dataframe
plt.plot(revenue["city"],revenue["revenue"],marker="o",mec="orange",mfc="orange") #plot graph to see revenue by order_date
plt.xlabel("City")
plt.ylabel("Revenue")
plt.title("Revenue by city")
```

**Revenue is dropping in Bangalore**

```python
# revenue trend by category
revenue = df.groupby("category")["revenue"].sum() #get revenue according to order_date
revenue = pd.DataFrame(revenue).reset_index() #convert series to dataframe
plt.plot(revenue["category"],revenue["revenue"],marker="o",mec="orange",mfc="orange") #plot graph to see revenue by order_date
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.title("Revenue by category")
```

**In Fashion Category Revenue got dropped**

```python
# count customers based on order_id (count of orders per customers)
df.groupby("customer_id")["order_id"].count()

```

**Most of the customers are ordering less**

```python
# Revenue trend for product_name
revenue = df.groupby("product_name")["revenue"].sum() #get revenue according to order_date
revenue = pd.DataFrame(revenue).reset_index() #convert series to dataframe
plt.plot(revenue["product_name"],revenue["revenue"],marker="o",mec="orange",mfc="orange") #plot graph to see revenue by order_date
plt.xlabel("Product name")
plt.ylabel("Revenue")
plt.title("Revenue by product name")
```

**Revenue is very less for shoes and watch**

# Insights:


1) Revenue got dropped in last two months (jun,jul)

2) Bangalore contributed more in revenue drop

3) Fashion category recorded less revenue

4) Repeated customers ordered less

5) Shoe and Watch sales declined significantly


# Recommendations:

1) Marketing in should be done in Bangalore

2) provide discounts for shoe and watch

3) Coupens,discounts,loyalty programs for repeated customers.
